# ML


```python
# label encoding 

from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
df.Gender = enc.fit_transform(df.Gender)
```

```python
df.drop('CustomerID', axis = 1, inplace = True)
```

```python
df.head()
```

Elbow Method to find the K(no.of.clusters) value
 **documentation info:sklearn.cluster.KMean**

```python
lst = []                                                    #go through the parameters in documentation
for i in range(1,11): # no of cluster values
  kmodel = KMeans(n_clusters = i, n_init = 15, max_iter = 500)  #  General Manupulation of k mean clustering for a particular k value
  kmodel.fit(df)                                                  # Fitting our data in to it
  lst.append(kmodel.inertia_)  # Kmodel.inertia = Wcss
```

```python
lst # values of Wcss 
```

```python
#plot of Wcss vs K =1 to 10
```

```python
plt.plot(range(1,11), lst, marker = 'o')
plt.grid()
```

```python
# k = 6 the first point that encounters less decrease
```

```python
kmodel = KMeans(n_clusters = 6) #  General Manupulation of k mean clustering for a particular k value
```

```python
kmodel.fit(df)
```

```python
prediction = kmodel.predict(df)
```

```python
kmodel.cluster_centers_          # cluster centers for each clusters out of 6 clusters
```

```python
len(kmodel.cluster_centers_)
```

```python
# as we are got 6 clusters , prediction is describing each id belongs to which cluster  ie; cluster 0  or cluster1 or cluster2 or cluster3 or cluster4 or cluster5
```

```python
df['clusters'] = prediction # adding new column to the df
```

```python
df.head(20)
```

```python
df.clusters.unique()
```

```python
color = np.array(['darkgray', 'lightsalmon', 'powderblue', 'red', 'yellow', 'black'])
sns.scatterplot(x = df['Annual Income (k$)'], y = df['Spending Score (1-100)'], c = color[kmodel.labels_], s = 50)
```

### Conclusions:


*   red color people have very low income but their expenditure is very high.
*   blue color people have high annual income and high expenditure.


*   yellow color people have very low income and are spending low.
*   pink color people have high annual income and are spending less.


*   black and gray color people are spending their whole income.
